```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import chaosmagpy as cp
```

```python
df_ambT = pd.read_csv('MAA paper cristiano files/AmbientTemper.dat', sep = '\s+', header = None, skiprows = 1, names = ['Time','AmbT'])
df_ambT['Time'] = cp.data_utils.timestamp(df_ambT['Time'])
df_ambT['Time'] = pd.to_datetime(df_ambT['Time'], format = '%Y-%m-%dd %H:%M:%S.%f')
df_ambT.index = df_ambT['Time']
df_ambT.pop('Time')
```

```python
df_elecT = pd.read_csv('MAA paper cristiano files/EletronicTemper.dat', sep = '\s+', header = None, skiprows = 1, names = ['Time','elecT'])
df_elecT['Time'] = cp.data_utils.timestamp(df_elecT['Time'])
df_elecT['Time'] = pd.to_datetime(df_elecT['Time'], format = '%Y-%m-%dd %H:%M:%S.%f')
df_elecT.index = df_elecT['Time']
df_elecT.pop('Time')
```

```python
df_sensT = pd.read_csv('MAA paper cristiano files/SensorTemper.dat', sep = '\s+', header = None, skiprows = 1, names = ['Time','sensT'])
df_sensT['Time'] = cp.data_utils.timestamp(df_sensT['Time'])
df_sensT['Time'] = pd.to_datetime(df_sensT['Time'], format = '%Y-%m-%dd %H:%M:%S.%f')
df_sensT.index = df_sensT['Time']
df_sensT.pop('Time')
```

```python
df_ambT
```

```python
fig, ax  = plt.subplots(4,1,figsize=(9,6),sharex=True)

ax[1].plot(df_elecT['elecT'],color = 'red')

ax[0].plot(df_sensT['sensT'],color = 'blue')

ax[2].plot(df_ambT['AmbT'],'green')

ax[3].plot(df_sensT['sensT']-df_sensT['sensT'].mean(),
           'blue',label = '$T_s$')
ax[3].plot(df_elecT['elecT'] - df_elecT['elecT'].mean(),
           'red',label = '$T_e$')

ax[3].plot(df_ambT['AmbT'] - df_ambT['AmbT'].mean(),'green',label = '$T_a$')

ax[3].plot((df_elecT['elecT']-df_elecT['elecT'].mean()) - (df_sensT['sensT'] - df_sensT['sensT'].mean()),
           color = 'orange',label = '$T_e$ - $T_s$')







ax[3].set_ylim()
ax[3].set_ylim(-5,8)
ax[3].legend(bbox_to_anchor=(1.17,4.5), loc='upper right')
fig.text(0.07, 0.5, 'Temperature (Â°C)', va='center', rotation='vertical', fontsize = 12)
for axis in ax:
    axis.set_xlim(np.datetime64('2020-01-10 00:00'),np.datetime64('2020-01-17 00:00'))
    axis.grid(alpha = 0.5)
    #axis.set_xlim(df_ambT.index[0],df_ambT.index[-1])
plt.savefig('fig10.jpeg', bbox_inches='tight', dpi = 1000)
plt.show()
```

```python
df_data = pd.read_csv('MAA paper cristiano files/MagneticSignal.dat', sep = '\s+', header = None, skiprows = 1, names = ['Time','F_geo','F_GSM',
                                                                                                                         'F_fit1','F_fit2','F_fit3'])
df_data['Time'] = cp.data_utils.timestamp(df_data['Time'])
df_data['Time'] = pd.to_datetime(df_data['Time'], format = '%Y-%m-%dd %H:%M:%S.%f')
df_data.index = df_data['Time']
df_data.pop('Time')
```

```python
df_data
```

```python
fig, axes = plt.subplots(5,1, figsize = (10,7), sharex= True)
ax1 = axes[1].twinx()
axes[0].plot(df_data['F_geo'] - df_data['F_GSM'], color = 'red',linewidth = 0.5)
axes[1].plot(df_data['F_fit1'] - df_data['F_GSM'],color = 'blue',linewidth = 0.5)
axes[1].set_ylim(-1,1)
axes[2].set_ylim(-1,1)
axes[3].set_ylim(-1,1)
ax1.plot(df_sensT['sensT'], color = 'black')


axes[0].set_ylabel('\u0394F\u2080 (nT)',fontsize = 10)
axes[1].set_ylabel('\u0394F\u2081 (nT)',fontsize = 10)
axes[2].set_ylabel('\u0394F\u2082 (nT)',fontsize = 10)
axes[3].set_ylabel('\u0394F\u2083 (nT)',fontsize = 10)
axes[4].set_ylabel('\u0394F (nT)',fontsize = 10)

ax2 = axes[2].twinx()
ax1.set_ylabel('$T_s$ (Â°C)')
ax2.set_ylabel('$T_e$-$T_s$ (Â°C)')
axes[2].plot(df_data['F_fit2'] - df_data['F_GSM'],color = 'green',linewidth = 0.5)
ax2.plot((df_elecT['elecT'] - df_sensT['sensT']) - (df_elecT['elecT'] - df_sensT['sensT']).mean(), color = 'black')
ax2.set_ylim(-8,8)
axes[3].plot(df_data['F_fit3'] - df_data['F_GSM'],color = 'orange',linewidth = 0.5)

axes[4].plot(df_data['F_geo'] - df_data['F_GSM'], color = 'red',linewidth = 0.5)
axes[4].plot(df_data['F_fit1'] - df_data['F_GSM'], color = 'blue',linewidth = 0.5)
axes[4].plot(df_data['F_fit2'] - df_data['F_GSM'],color = 'green',linewidth = 0.5)
axes[4].plot(df_data['F_fit3'] - df_data['F_GSM'],color = 'orange',linewidth = 0.5)
#plt.xlim(np.datetime64('2020-01-14 00:00'),np.datetime64('2020-01-15 00:00 '))
#fig.text(0.07, 0.5, 'Temperature (Â°C)', va='center', rotation='vertical', fontsize = 12)

for ax in(axes.flatten()):
    ax.set_xlim(np.datetime64('2020-01-10 00:00'),np.datetime64('2020-01-17 00:00'))
    ax.grid(alpha = 0.5)
    
plt.savefig('fig11.jpeg', bbox_inches='tight', dpi = 1000)
plt.show()
```
