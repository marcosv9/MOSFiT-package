```python
def download_data_INTERMAGNET(Year, Months, files = None):
    
    '''
    Download observatory files from Intermagnet FTP server and save in
    a specific directory.
    
    Datatype must be: 'D' for Definitive or 'QD' quasi-definitive
    
    Year must be informed as '2021', for example.
    
    Months must be a list, for example - ['01','02','03']
    
    files must be an obs IAGA code or None, if None, files for all the observatories will
    be downloaded.
    
    example of use - mvs.download_data_INTERMAGNET('QD', '2021', ['07','08','09'], files = None)
    '''
    List_Months = ['01','02','03','04','05','06','07','08','09','10','11','12']
    ftp = ftplib.FTP('seismo.nrcan.gc.ca')
    ftp.login('anonymous', 'email@email.com')
    
    
    if Months == None:
        Months = List_Months
        
    for month in Months:
        directory = 'Dados OBS/' + Year + '/' + month
        #print(directory)
        #if datatype == 'QD':
        path_qd = 'intermagnet/minute/quasi-definitive/IAGA2002/' + Year + '/' + month
            
        #if datatype == 'D':
        path_d = 'intermagnet/minute/definitive/IAGA2002/' + Year + '/' + month
        
        if files == None:
            #ftp connection for def files
            try:
                ftp = ftplib.FTP('seismo.nrcan.gc.ca')
                ftp.login('anonymous', 'email@email.com')
                ftp.cwd(path_d)
                filenames = ftp.nlst('*') # get filenames within the directory
                filenames.sort()
            except:
                filenames = []
            
            #ftp connection for QD files
            try:
                ftp2 = ftplib.FTP('seismo.nrcan.gc.ca')
                ftp2.login('anonymous', 'email@email.com')
                ftp2.cwd(path_qd)
                filenamesqd = ftp2.nlst('*') # get filenames within the directory
                filenamesqd.sort()
            except:
                filenamesqd = []
            
            if filenames != [] and filenamesqd != []:
                
                for filename, filenameqd in zip(filenames, filenamesqd):
                    if os.path.isfile(directory +'/' + filename) == True and os.path.isfile(directory +'/' + filenameqd) == False:
                        
                        print('File ' + filename  + ' already exist!')
                        
                    if os.path.isfile(directory +'/' + filename) == False and os.path.isfile(directory +'/' + filenameqd) == True:
                        
                        os.remove(directory +'/' + filenameqd) 
                        print('File ' + filenameqd  + ' removed!')
                        
                        pathlib.Path(directory).mkdir(parents=True, exist_ok=True)   
                        local_filename = os.path.join(directory, filename)
                        file = open(local_filename, 'wb')
                        ftp.retrbinary('RETR '+ filename, file.write)
                        print('File ' + filename  + ' downloaded!')
                        
                    if os.path.isfile(directory +'/' + filename) == True and os.path.isfile(directory +'/' + filenameqd) == True:    
                        
                        os.remove(directory +'/' + filenameqd) 
                        print('Duplicate file. File ' + filenameqd  + ' removed!')
                        
                        pathlib.Path(directory).mkdir(parents=True, exist_ok=True)
                        local_filename = os.path.join(directory, filename)
                        file = open(local_filename, 'wb')
                        ftp.retrbinary('RETR '+ filename, file.write)
                        print('File ' + filename  + ' downloaded!')
            
            if filenames == [] and filenamesqd != []:
                    
                for filenameqd in filenamesqd:
                        
                    if os.path.isfile(directory +'/' + filenameqd) == True:
                        print('File ' + filenameqd  + ' already exist!')
                    else:
                        pathlib.Path(directory).mkdir(parents=True, exist_ok=True)
                        local_filename = os.path.join(directory, filenameqd)
                        file = open(local_filename, 'wb')
                        ftp2.retrbinary('RETR '+ filenameqd, file.write)
                        print('File ' + filenameqd  + ' downloaded!')
                            
            if filenames != [] and filenamesqd == []:
                pathlib.Path(directory).mkdir(parents=True, exist_ok=True)
                for filename in filenames:
                    
                    if os.path.isfile(directory +'/' + filename) == True:
                        
                        print('File ' + filename  + ' already exist!')
                        
                    else:
                        pathlib.Path(directory).mkdir(parents=True, exist_ok=True)
                        local_filename = os.path.join(directory, filename)
                        file = open(local_filename, 'wb')
                        ftp.retrbinary('RETR '+ filename, file.write)
                        print('File ' + filename  + ' downloaded!')
            ftp.quit()
            print('Disconnected from INTERMAGNET Ftp server!') 

        if files != None:
            for file in files:
                    
                #ftp connection for def files
                try:
                    ftp = ftplib.FTP('seismo.nrcan.gc.ca')
                    ftp.login('anonymous', 'email@email.com')
                    ftp.cwd(path_d)
                    filenames = ftp.nlst(file.lower() + '*') # get filenames within the directory
                    filenames.sort()
                    
                except:
                    filenames = []
                #print(filenames)
                
                #ftp connection for QD files
                try:
                    ftp2 = ftplib.FTP('seismo.nrcan.gc.ca')
                    ftp2.login('anonymous', 'email@email.com')
                    ftp2.cwd(path_qd)
                    filenamesqd = ftp2.nlst(file.lower() + '*') # get filenames within the directory
                    filenamesqd.sort()

                except:
                    filenamesqd = []
                #print(filenamesqd)
                    #print('List of files that will be downloaded')
                    #for filename in filenames:
                    #    print(filename)
                    
                
                if filenames != [] and filenamesqd != []:   
                    for filename, filenameqd in zip(filenames, filenamesqd):
                        if os.path.isfile(directory +'/' + filename) == True and os.path.isfile(directory +'/' + filenameqd) == False:
                            
                            print('File ' + filename  + ' already exist!')
                            
                        if os.path.isfile(directory +'/' + filename) == False and os.path.isfile(directory +'/' + filenameqd) == True:
                            
                            os.remove(directory +'/' + filenameqd) 
                            print('File ' + filenameqd  + ' removed!')
                            
                               
                            local_filename = os.path.join(directory, filename)
                            file = open(local_filename, 'wb')
                            ftp.retrbinary('RETR '+ filename, file.write)
                            print('File ' + filename  + ' downloaded!')
                            
                        if os.path.isfile(directory +'/' + filename) == True and os.path.isfile(directory +'/' + filenameqd) == True:    
                            
                            os.remove(directory +'/' + filenameqd) 
                            print('Duplicate file. File ' + filenameqd  + ' removed!')
                            
                            
                            local_filename = os.path.join(directory, filename)
                            file = open(local_filename, 'wb')
                            ftp.retrbinary('RETR '+ filename, file.write)
                            print('File ' + filename  + ' downloaded!')
                            
                        if os.path.isfile(directory +'/' + filename) == False and os.path.isfile(directory +'/' + filenameqd) == False:    
                                                   
                            local_filename = os.path.join(directory, filename)
                            file = open(local_filename, 'wb')
                            ftp.retrbinary('RETR '+ filename, file.write)
                            print('File ' + filename  + ' downloaded!')
                            
                
                if filenames == [] and filenamesqd != []:
                        
                    for filenameqd in filenamesqd:
                            
                        if os.path.isfile(directory +'/' + filenameqd) == True:
                            print('File ' + filenameqd  + ' already exist!')
                        else:
                            local_filename = os.path.join(directory, filenameqd)
                            file = open(local_filename, 'wb')
                            ftp2.retrbinary('RETR '+ filenameqd, file.write)
                            print('File ' + filenameqd  + ' downloaded!')
                                
                if filenames != [] and filenamesqd == []:
                    
                    for filename in filenames:
                        
                        if os.path.isfile(directory +'/' + filename) == True:
                            
                            print('File ' + filename  + ' already exist!')
                            
                        else:
                            
                            local_filename = os.path.join(directory, filename)
                            file = open(local_filename, 'wb')
                            ftp.retrbinary('RETR '+ filename, file.write)
                            print('File ' + filename  + ' downloaded!')
                            
                            
                        
                        
                     
            ftp.quit()
            print('Disconnected from INTERMAGNET Ftp server!') 
```

```python

import sys
sys.path.insert(0, 'C:/Users/marco/Downloads/Thesis_notebooks/SV_project')
import utilities_tools as utt
import chaosmagpy as cp
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%config InlineBackend.figure_format ='retina'
from glob import glob
import glob
import matplotlib.gridspec as gridspec
import main_functions as mvs
import data_processing_tools as dpt
import utilities_tools as utt
#import pwlf
%matplotlib inline
import pathlib
from pandas.tseries.frequencies import to_offset
from datetime import datetime,date
import ftplib
import os

import timeit
from datetime import datetime 
import pwlf

#import missingno as missingno

def download_data_INTERMAGNET(datatype,
                              year,
                              months,
                              station = None
                             ):
    
    '''
    Download observatory files from Intermagnet FTP server and save in
    a specific directory.
    
    Datatype must be: 'D' for Definitive or 'QD' quasi-definitive
    
    Year must be informed as '2021', for example.
    
    Months must be a list, for example - ['01','02','03']
    
    files must be an obs IAGA code or None, if None, files for all the observatories will
    be downloaded.
    
    example of use - mvs.download_data_INTERMAGNET('QD', '2021', ['07','08','09'], files = None)
    '''
    
    assert datatype in ['QD','D'], 'datatype must be QD or D'
    
    df_IMOS = pd.read_csv('Data/Imos informations/IMOS_INTERMAGNET.txt',
                           skiprows = 1,
                           sep = '\s+',
                           usecols=[0, 1, 2, 3],
                           names = ['Imos', 'Latitude', 'Longitude', 'Elevation'],
                           index_col= ['Imos'])
    
    List_Months = ['01', '02', '03',
                   '04', '05', '06',
                   '07', '08', '09',
                   '10', '11', '12'
                   ]
    
    ftp = ftplib.FTP('seismo.nrcan.gc.ca')
    ftp.login('anonymous', 'email@email.com')
    
    if months == None:
        months = List_Months
        
    if station == None:
        
        for month in months:
            directory = 'Dados OBS/' + year + '/' + month
            print(directory)
            if datatype == 'QD':
                path = f'intermagnet/minute/quasi-definitive/IAGA2002/{year}/{month}'
                
            if datatype == 'D':
                path = f'intermagnet/minute/definitive/IAGA2002/{year}/{month}'
                
            for station in df_IMOS.index[0:150]:    
                
                ftp = ftplib.FTP('seismo.nrcan.gc.ca')
                ftp.login('anonymous', 'email@email.com')
                ftp.cwd(path)
                try:
                    filenames = ftp.nlst(f'{station.lower()}*') # get filenames within the directory
                    filenames.sort()
                    pathlib.Path(directory).mkdir(parents=True, exist_ok=True)               
                    for filename in filenames:
                        
                        if os.path.exists(f'{directory}/{filename}') == False: 
                            print('File ' + filename  + ' downloaded!')   
                            local_filename = os.path.join(directory, filename)
                            file = open(local_filename, 'wb')
                            ftp.retrbinary('RETR '+ filename, file.write)
                            directory_files = glob.glob(f'{directory}/{filename[0:11]}*')
                            if len(directory_files) > 1:
                                os.remove(f'{directory}/{filename[0:11]}qmin.min')
                                print(f'QD file {filename[0:11]}qmin.min removed, there is a definitive data.')
                        else:
                            print(f'file {filename} already exist')
                            pass
                        
                    ftp.quit()
                except:
                    pass
    if station != None:
        for station in station:
            print(station)
            for month in months:
                directory = 'Dados OBS/' + year + '/' + month
                print(directory)
                if datatype == 'QD':
                    path = f'intermagnet/minute/quasi-definitive/IAGA2002/{year}/{month}'
                
                if datatype == 'D':
                    path = f'intermagnet/minute/definitive/IAGA2002/{year}/{month}'
                    
                ftp = ftplib.FTP('seismo.nrcan.gc.ca')
                ftp.login('anonymous', 'email@email.com')
                ftp.cwd(path)
                
                try:
                    filenames = ftp.nlst(f'{station.lower()}*') # get filenames within the directory
                    filenames.sort()
                    pathlib.Path(directory).mkdir(parents=True, exist_ok=True)               
                    for filename in filenames:
                        
                        if os.path.exists(f'{directory}/{filename}') == False: 
                            print('File ' + filename  + ' downloaded!')   
                            local_filename = os.path.join(directory, filename)
                            file = open(local_filename, 'wb')
                            ftp.retrbinary('RETR '+ filename, file.write)
                            directory_files = glob.glob(f'{directory}/{filename[0:11]}*')
                            if len(directory_files) > 1:
                                os.remove(f'{directory}/{filename[0:11]}qmin.min')
                                print(f'QD file {filename[0:11]}qmin.min removed, there is a definitive data.') 
                        else:
                            print(f'file {filename} already exist')
                            pass
                        
            
                except:
                    pass
            
        ftp.quit()
        print('Disconnected from INTERMAGNET Ftp server!')
```
